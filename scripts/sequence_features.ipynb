### 에피소드 특성
  총: 15,435개
  평균 길이: 23.1개 액션
  평균 지속: 43.8초
  Pass 비율: 48.9%
  
  해석:

    → 한 에피소드 = 약 44초간 23개 액션
    → 절반이 패스 (짧은 패스 위주 빌드업)
    → 평균 2초마다 액션 발생 (빠른 템포!)

### 액션 시퀀스 패턴

  1. Pass -> Pass (75,541번)
  2. Carry -> Pass(71,354번)
  3. Pass -> Carry(62,321번)

  - K리그 스타일 :
    - Pass, Carry, Pass 순환
    - 짧은 패스 + 드리블 조합

### 시간/템포 특성
  - 템포 :
    - 평균 액션 간격 : 1.89초
    - 빠른 템포 : 229,640개 (64%)
    - 느린 템포 : 4,303개 (1%)
  
  - K리그는 빠른 템포(2초 이내 빠른 전개 및 압박 축구 구사)

### 공간 이동
  - 이동 거리:
    - 액션 간 평균: 18.51m
    - 에피소드 누적 전진: 73.74m
  
  - 공격 전개:
    - 한 공격 당 약 74m 전진
    - 액션 당 18m씩 이동
    - 3-4번 액션으로 필드 절반 진행

### 연속 패스
  - 최대 연속 패스: 17개 (빌드업 중심 전술)
